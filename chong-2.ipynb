```python
from fastai.vision.all import *
```

```python
path = Path('images')
```

```python
item_tfms = Resize(224)
```

```python
batch_tfms = aug_transforms()
```

```python
dls = ImageDataLoaders.from_folder(
    path,
    train = 'train',
    valid = 'val',
    seed = 42,
    item_tfms = item_tfms,
    batch_tfms = batch_tfms,
    batch_size = 4)
```

```python
dls.show_batch()
```

```python
import torchvision.models as models
```

```python
from fastai.vision.all import *

arch = models.vgg16_bn
learn = cnn_learner(dls, arch, pretrained=True, metrics=accuracy)
```

```python
from fastai.callback.tracker import EarlyStoppingCallback, SaveModelCallback
```

```python
early_stop_cb = EarlyStoppingCallback
```

```python
save_model_cb = SaveModelCallback()
```

```python
learn.fine_tune(100, 3e-3, cbs=[early_stop_cb, save_model_cb])
```

```python
learn.show_results()
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
interp.print_classification_report()
```

```python
interp.plot_top_losses(9, figsize=(15,10))
```

```python
test_dl = dls.test_dl(get_image_files(path/'test'),
                      with_labels=True)
```
